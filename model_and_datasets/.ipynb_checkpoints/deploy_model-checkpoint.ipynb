Melakukan pengecekan pada data, untuk menemukan nilai mising value pada masing-masing kolom di datasets

```python
misingness = df.isnull().sum().sum()/df.count().sum()
print('Overall Misiness pada data adalah : {:.2f}%'.format(misingness*100))
```

```python
def missing_table(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    mising_data = pd.concat([total, percent], axis=1, keys=['Total','Percent'])
    return mising_data

missing_table(df).head(10)
```
